```python
#CAN WE CHANGE THE SATISFACTION RATING INTO SOMETHING MORE APPROPRIATE
laptop_purchase_data_india['Satisfaction_Rating'] = laptop_purchase_data_india['Satisfaction_Rating'].replace(1,'Very_Poor')
laptop_purchase_data_india['Satisfaction_Rating'] = laptop_purchase_data_india['Satisfaction_Rating'].replace(2,'Poor')
laptop_purchase_data_india['Satisfaction_Rating'] = laptop_purchase_data_india['Satisfaction_Rating'].replace(3,'Good')
laptop_purchase_data_india['Satisfaction_Rating'] = laptop_purchase_data_india['Satisfaction_Rating'].replace(4,'Very_Good')
laptop_purchase_data_india['Satisfaction_Rating'] = laptop_purchase_data_india['Satisfaction_Rating'].replace(5,'Excellent')
```

```python
#HERE, WE WANNA DO SOME GROUPBY AND AGGREGATIONS WE MIGHT BE VISUALIZATION AS WE GO ON
Count_Genders = laptop_purchase_data_india.groupby('Gender')['Customer_ID'].count()
Count_Genders 
```

```python
Count_Genders.plot.pie(title = 'Count_of_Genders', color = 'black' )
```

```python
Total_Customers_by_Payment_Mode = laptop_purchase_data_india.groupby('Payment_Mode')['Customer_ID'].count()
Total_Customers_by_Payment_Mode
```

```python
Total_Customers_by_Payment_Mode.plot(kind = 'bar', title = 'Total_Customers_by_Payment_Mode', ylabel = 'Count',
                                     xlabel = 'Payment_Mode', color = 'black', stacked = True)
```

```python
Count_of_Customers_by_Income_Level_And_Gender = laptop_purchase_data_india.groupby(['Gender', 'Income_Level'])['Customer_ID'].count()
Count_of_Customers_by_Income_Level_And_Gender
```

```python
Count_of_Customers_by_Income_Level_And_Gender.plot.barh(title = 'Count_of_Customers_by_Income_Level_And_Gender', ylabel = 'Gender/Income_Levels',
                                                        xlabel = 'Count', color = 'black', stacked = True)
```

```python
Count_laptop_brands_purchased_and_Purchase_mode = laptop_purchase_data_india.groupby(['Brand_Purchased', 'Purchase_Mode'])['Customer_ID'].count()
Count_laptop_brands_purchased_and_Purchase_mode
```

```python
Count_laptop_brands_purchased_and_Purchase_mode.plot(kind = 'bar', title = 'Count_laptop_brands_purchased_and_Purchase_mode', ylabel = 'count',
                                                     xlabel = 'Brand_Name_And_Purchased_TYpe', color = 'black', stacked = True)
```

```python
Frequency_Purchase = laptop_purchase_data_india.groupby('Purchase_Frequency')['Customer_ID'].count()
Frequency_Purchase
```
